```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as py
```

```python
def avg_data_2013():
    average=[]
    for row in pd.read_csv('Data/AQI/aqi2013.csv', chunksize=24):
        add = 0
        for index,eachRow in row.iterrows():
            if type(eachRow['PM2.5']) is int or type(eachRow['PM2.5']) is float:
                add = add + eachRow['PM2.5']
            elif type(eachRow['PM2.5']) is str:
                if eachRow['PM2.5']!='NoData' and eachRow['PM2.5']!='PwrFail' and eachRow['PM2.5']!='---' and eachRow['PM2.5']!='InVld':
                    add= add+ float(eachRow['PM2.5'])
        average.append(add/24)
    return average
```
